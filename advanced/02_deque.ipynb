### `deque` in Python

`deque` is double ended queue, which supports append and pop elements at both the ends. `deque` is implemented using doulbe linked list, which is very efficient for insert and delete elements at O(1).

```python
from collections import deque
```

```python
empty_deque = deque()
empty_deque
```

```python
# deque with elements
prime_numbers = deque([3,5,7,11])
print(prime_numbers)
```

```python
# Append elements at the right end
prime_numbers.append(13)
prime_numbers
```

```python
# Append element at the left end
prime_numbers.appendleft(2)
prime_numbers
```

```python
# pop element from the right end
prime_numbers.pop()
```

```python
# pop element from the left end
prime_numbers.popleft()
```

```python
# Final elements in the prime_numbers
prime_numbers
```
